
## Problem Statement

### 2. Plot a bar graph no. of matches won by each team in the year 2011

```python
#Task2 Setting Functions Filtering Data and setting a Dataframe
def filt(dp,x,y):
    li=[]
    count=0
    columns=dp[x]
    for i in columns:
        if (i==y):
            li.append(dp.values[count])
        count+=1
    return li
fdf=pd.DataFrame(filt(df,"Season_Year",2011),columns=df.columns)
fdf
```

```python
#Counting Teams And No. Of Wins Per Team
teams=list(fdf["match_winner"].unique())
w_count={}
for i in teams:
    w_count[i]=0
for i in fdf["match_winner"]:
    w_count[i]+=1
del w_count[np.nan]
```

```python
#Bar Graph
index=np.arange(0,len(w_count))
plt.barh(index ,w_count.values() , color="gray")
plt.yticks(index,teams)
plt.xlabel('No of Wins', fontsize=20)
plt.ylabel('Teams', fontsize=20)
plt.show()
```

### 3. Print ManOfMatch count of each player in Hyderabad stadium

```python
#Task 3 Filtering The Data Using Above Function
fdfc=pd.DataFrame(filt(df,"City_Name","Hyderabad"),columns=df.columns)
fdfc
```

```python
#Counting Players And No. Of Man Of Match Per Player
mom=list(fdfc["ManOfMach"].unique())
mom_count={}
for i in mom:
    mom_count[i]=0
for i in fdfc["ManOfMach"]:
    mom_count[i]+=1
for i in mom_count:
    print("Name: ",i,"\n","Matches: ",mom_count[i],end="\n\n")
```
